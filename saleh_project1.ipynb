```python
task_list=[]

def add_task():
    tsub=input("Enter task subject:\n")
    tdesc=input("Enter task description:\n")
    tdead=int(input("Enter task deadline (in minutes)\n"))
    Chekced=False
    task_list.append({'subject':tsub, 'description': tdesc,
                       'deadline':tdead, 'finished?':Chekced})
    print(f'{tsub} Task added successfully')

def remove_task():
    t=task_list.copy()
    tname=input("Enter the task subject:\n")
    if len(task_list) == 0:
        print("The To-Do list is empty")
    else:
        for i in range(len(t)):
            if tname == t[i]['subject']:
                del task_list[i]
                print(f'The Task {tname} deleted form the list successfully')
                break
        if len(t)==len(task_list):
            print(f'The task {tname} not found')
def achive_task():
    tsub=input("which task you achived?\n")
    for i in range(len(task_list)):
        if tsub == task_list[i]['subject']:
            task_list[i]['finished?']=True
            print("task checked")
            break

def display_task():
    if not task_list:
        print('No tasks to display')
    else:
        print('The task list is: ')
        for i in task_list:
            print(i)
        

def ui():
    print('**','*'*31,'**\n*',sep='',end='')
    print("Welcome to the To-Do list program",end='')
    print("*\n",'**','*'*31,'**\n',sep='')
    print("Enter the number of opreation you want to do:")
    print("1. Add Task to To-Do List")
    print("2. Remove Task From To-Do List")
    print("3. Display To-Do List")
    print("4. to check task")
    print("5. Quit")
```

```python
while True:
    ui()
    try:
        Choice=int(input())
        if Choice==1:
            add_task()
        elif Choice==2:
            remove_task()
        elif Choice==3:
            display_task()
        elif Choice==4:
            achive_task()
        elif Choice==5:
            break
        else:
            print("you enterd wrong number")
    except Exception as e:
        print(e)
```
