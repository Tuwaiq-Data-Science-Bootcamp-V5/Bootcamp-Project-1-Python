```python
students = {}

 
def add (name, age, grade): 
    students[name] = {'Name':name,'Age': age, 'Grade': grade}
    
    

def get_student_details(name):
    if name in students:
        return students[name]
    else:
        return 'Student not found!'

    
    
def filter_students_by_grade(min_grade):
    filtered_students = {name: details for name, details in students.items() if details['Grade'] >= min_grade }
    return (filtered_students)


to_lower = lambda name: name.lower()



 
while True:
    print("Student Management System")
    print("1. Add Student")
    print("2. Get Student Details")
    print("3. Filter Students by Grade")
    print("4. Exit")
    enter = input("Enter your choice (1-4): ")

    if enter == '1':
        name = input("Enter student name: ")
        name = to_lower(name)
        age = int(input("Enter student age: "))
        grade = float(input("Enter student grade: "))
        add (name, age, grade)
        print("Student added successfully!")
        
        
    elif enter == '2':
        name = input("Enter student name: ")
        details = get_student_details(name)
        print("Student Details:")
        print("Name:", name)
        if isinstance(details, dict):
            print("Age:", details['Age'])
            print("Grade:", details['Grade'])
        else:
            print(details)
            
            
    elif enter == '3':
        min_grade = float(input("Enter the minimum grade: "))
        filtered_students = filter_students_by_grade(min_grade)
        print("Filtered Students:")
        if len(filtered_students) > 0:
            for name, details in filtered_students.items():
                print("Name:", name)
                print("Age:", details['Age'])
                print("Grade:", details['Grade'])
                print("-----------")
                
                
        else:
            print("No students meet the minimum grade.")
    elif enter == '4':
        print("Exit")
        break
    else:
        print("Invalid choice! Please try again.")
```
